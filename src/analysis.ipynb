```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
training_sizes = [0.2, 0.4, 0.6, 0.8]
results = []
for size in training_sizes:
    for stage in ['pre', 'post']:
        filename = f'../results/Chase_{stage}_train{round(size, 1)}.txt'
        if os.path.exists(filename):
            with open(filename, 'r') as file:
                loaded_results = [float(line.strip()) for line in file]
            results.append(loaded_results)
fig, ax = plt.subplots(figsize=[10, 8])
# Boxplot
colors = ['lightblue', 'lightgreen'] * len(training_sizes)
positions = []
for i, size in enumerate(training_sizes):
    offset = i*3  # for every training size, shift by 3 positions
    positions += [offset+1, offset+2]

boxes = ax.boxplot(results, vert=True, patch_artist=True, positions=positions)
# boxes = ax.boxplot(results, vert=True, patch_artist=True, labels=labels)

for patch, color in zip(boxes['boxes'], colors):
    patch.set_facecolor(color)
    
ax.set_xticks(np.arange(1.5, 3*len(training_sizes), 3))  # place tick in the middle of each grouped boxplots
ax.set_xticklabels([str(size) for size in training_sizes])
ax.set_title('Comparison of Machine Learning Results (100 Experiments)\nby Training Size and Stage')
ax.set_xlabel('Training Size and Stage')
ax.set_ylabel('Recalls')

medians = [np.median(dataset) for dataset in results]
for position, median in zip(positions, medians):
    ax.text(position, median + 0.01, f'{median:.2f}', horizontalalignment='center', verticalalignment='bottom') 

# Display legend
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], color='lightblue', label='pre'),
                   Line2D([0], [0], color='lightgreen', label='post')]
ax.legend(handles=legend_elements, loc='upper left')

plt.tight_layout()
plt.show()
```
